## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
# Based on the boxplot of the column "Age", where do you see the outliers?

import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

# Calculate the lower and upper bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers below Q1 and above Q3
outliers_below_q1 = df[df['Age'] < lower_bound]
outliers_above_q3 = df[df['Age'] > upper_bound]

# Count the number of outliers below Q1 and above Q3
num_outliers_below_q1 = outliers_below_q1.shape[0]
num_outliers_above_q3 = outliers_above_q3.shape[0]

# Print the counts
print(f"The number of outliers below Q1 in the 'Age' column is: {num_outliers_below_q1}")
print(f"The number of outliers above Q3 in the 'Age' column is: {num_outliers_above_q3}")
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Extract the column related to the databases respondents wish to learn next year
databases_to_learn = df.loc[1:, 'DatabaseDesireNextYear'].dropna()

# Split the databases by ';' and create a series of all databases
all_databases_to_learn = databases_to_learn.str.split(';').explode()

# Count the occurrences of each database
database_counts = all_databases_to_learn.value_counts()

# Get the top 5 databases
top_5_databases = database_counts.head(5)

# Create the pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Filter the DataFrame for the age group 30 to 35
df_age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median of WorkWeekHrs and CodeRevHrs for this age group
median_work_week_hrs = df_age_group['WorkWeekHrs'].median()
median_code_rev_hrs = df_age_group['CodeRevHrs'].median()

# Data for the bar chart
categories = ['WorkWeekHrs', 'CodeRevHrs']
medians = [median_work_week_hrs, median_code_rev_hrs]

# Create the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(categories, medians, color=['blue', 'green'])

# Add labels and title
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
ax.set_xlabel('Categories')
ax.set_ylabel('Median Hours')
ax.legend(['Median Hours'])

# Show the chart
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Filter the DataFrame to include only respondents with ages from 45 to 60
df_45_60 = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_converted_comp = df_45_60.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the median ConvertedComp for ages 45 to 60
plt.figure(figsize=(10, 6))
sns.lineplot(data=median_converted_comp, x='Age', y='ConvertedComp', marker='o')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp (USD)')
plt.show()
```

```python
# Based on the Line Chart of Age and median ConvertedComp for the age group 25-30, what kind of correlation do you see between Age and median ConvertedComp?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Filter the DataFrame to include only respondents with ages from 25 to 30
df_25_30 = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

# Calculate the median ConvertedComp for each age within the 25-30 age group
median_converted_comp = df_25_30.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the median ConvertedComp for ages 25 to 30
plt.figure(figsize=(10, 6))
sns.lineplot(data=median_converted_comp, x='Age', y='ConvertedComp', marker='o')
plt.title('Median ConvertedComp for Ages 25 to 30')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp (USD)')
plt.show()
```

```python
# Of Respondents who work with MySQL, what other databases do they work with?

import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Extract the column related to the databases respondents currently work with
current_technologies = df['DatabaseWorkedWith'].dropna()

# Filter respondents who work with MySQL and split the databases
mysql_users = current_technologies[current_technologies.str.contains('MySQL')]

# Create a DataFrame to list respondents and their databases
mysql_df = mysql_users.str.split(';').explode().reset_index()
mysql_df.columns = ['Respondent', 'Database']

# Filter out MySQL to see what other databases respondents work with
other_databases = mysql_df[mysql_df['Database'] != 'MySQL']

# Count the occurrences of other databases
other_database_counts = other_databases['Database'].value_counts()

# Display the other databases
print("Other databases respondents work with along with MySQL:")
print(other_database_counts)
```
